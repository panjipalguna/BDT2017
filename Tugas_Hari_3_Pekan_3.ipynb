```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(16,8), nrows=1, ncols=3,sharey=True)
ax[0].plot(x,tan,color="b")
ax[0].set_title("Grafik tan x")
ax[1].plot(x,cos,color="r")
ax[1].set_title("Grafik sin x")
ax[2].plot(x,sin,color="g")
ax[2].set_title("Grafik cos x")
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig,ax = plt.subplots(figsize=(16,8),sharey=True)
ax.plot(x,y)
ax1=ax.inset_axes([-0.1,-0.1,0.3,0.3])
ax1.plot(x,y,'b',x,y2,'r')
ax1.set_title("Grafik CosX^2")
ax2=ax.inset_axes([-0.1,0.9,0.3,0.3])
ax2.plot(x,y,'b',x,y5,'y')
ax2.set_title("Grafik CosX^5")
ax3=ax.inset_axes([0.9,0.9,0.3,0.3])
ax3.plot(x,y,'b',x,y4,'black')
ax3.set_title("Grafik CosX^4")
ax4=ax.inset_axes([0.9,-0.1,0.3,0.3])
ax4.plot(x,y,'b',x,y3,'g')
ax4.set_title("Grafik CosX^3")
plt.title("Grafik CosX")
plt.show()
```
